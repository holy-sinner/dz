```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data['robot'] = pd.Series(['1' if i == 'robot' else '0' for i in data['whoAmI']])
data['human'] = pd.Series(['1' if i == 'human' else '0' for i in data['whoAmI']])
data
```
